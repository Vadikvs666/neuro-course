Решим по методу градиентного спуска
$$
a_n = a_{n-1} - f'(x)*h
$$

```python
import numpy as np
import sympy as sp
import matplotlib
import matplotlib.pyplot as plt
from sympy.plotting import plot

h=0.1
count=2




def diffXX(x):
    return 2*x

x = sp.Symbol('x', real=True)
f =  x * x 
df = f.diff(x)
b = x - df*x*h

print ("Производная: ")
print(df)

ai=2


for  a in range(count):
    ai=ai-diffXX(ai)*h
    print(ai)



pl1 = plot(f,(x, -3, 3),show = False)
pl2 = plot(df,(x, -3, 3),show = False)
pl3 = plot(b,(x, -2, 2),show = False)
pl1.append(pl2[0])
pl1.append(pl3[0])
pl1.show()
```
